```python
from datetime import datetime
import matplotlib.pyplot as plt
from GDELTCall import GDELT
```

```python
gdelt = GDELT("sanctions russia", start=datetime.strptime('2019-01-01', '%Y-%m-%d'), end=datetime.now(), source_country="AU")
```

```python
df = gdelt.get_timeline()
df
```
